```python
import matplotlib.pyplot as plot 
x=[2,6]
y=[4,9]
#this is for graph plotting using 2 cordinates
#here x.label and y.label is used to describe the cordinates
#hete legend() is used to describe particular line 
plot.xlabel("time")
plot.ylabel("speed")
plot.plot(x,y,label="river")
plot.legend()
plot.grid()
plot.show()
```
